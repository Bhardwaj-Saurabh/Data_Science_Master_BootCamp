**Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

**Answer:**

Seaborn is a popular Python data visualization library based on Matplotlib that provides a high-level interface for creating attractive and informative statistical graphics. Here are five common plots that can be created using Seaborn along with their uses:

Seaborn scatter plot: A scatter plot displays the relationship between two continuous variables. It is used to identify trends, patterns, or correlations between two variables.

Seaborn line plot: A line plot displays the relationship between two continuous variables as a line connecting the data points. It is used to visualize trends or patterns over time or any other continuous variable.

Seaborn bar plot: A bar plot displays the distribution of a categorical variable using bars. It can be used to compare the values of a categorical variable across different categories.

Seaborn histogram: A histogram displays the distribution of a single continuous variable by dividing it into bins and representing the frequency of data points in each bin. It is used to visualize the distribution of data and identify patterns or trends.

Seaborn box plot: A box plot displays the summary statistics of a continuous variable, such as the median, quartiles, and outliers, in a box-and-whisker format. It is used to identify skewness, variability, and outliers in data.

**Que 2: Load the "fmri" dataset using the load_dataset function of seaborn.**

Plot a line plot using

x ="timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.


```python
import seaborn as sns

# Load the "fmri" dataset
df = sns.load_dataset("fmri")

# Plot line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=df)

```

**Que 3: Load the "titanic" dataset using the load_dataset function of seaborn.**

Plot two box plots using 

x ='pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
df = sns.load_dataset("titanic")

# Create a subplot with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Box plot for 'age'
sns.boxplot(x='pclass', y='age', data=df, ax=axes[0])
axes[0].set_title('Box Plot of Age by Pclass')

# Box plot for 'fare'
sns.boxplot(x='pclass', y='fare', data=df, ax=axes[1])
axes[1].set_title('Box Plot of Fare by Pclass')

# Show the plots
plt.show()
```

**Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.**


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
df = sns.load_dataset("diamonds")

# Plot histogram with 'price' column and 'cut' column as hue
sns.histplot(data=df, x='price', hue='cut')
plt.title('Histogram of Price with Cut as Hue')
plt.show()
```

**Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.**


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
df = sns.load_dataset("iris")

# Plot pair plot with 'species' column as hue
sns.pairplot(data=df, hue='species')
plt.title('Pair Plot of Iris Dataset with Species as Hue')
plt.show()
```

**Que 6: Use the "flights" dataset from seaborn to plot a heatmap.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
df = sns.load_dataset("flights")

# Pivot the data to create a matrix of month vs year
pivot_df = df.pivot("month", "year", "passengers")

# Create a heatmap
sns.heatmap(pivot_df, cmap="YlGnBu", annot=True, fmt="d", cbar=True)
plt.title("Flights Dataset Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
