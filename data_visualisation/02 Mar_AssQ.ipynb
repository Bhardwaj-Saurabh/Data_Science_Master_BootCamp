Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

**ANSWER:**

Matplotlib is a python library which helps to plot difficult graph to visualise data. It is used to understand the data better. Various statistical analysis also can be made using matplotlib. 

Followings are few plot which can be plotted using matplotlib:

- Scatter plot
- box plot
- violin plot
- histogram plot
- heatmap
