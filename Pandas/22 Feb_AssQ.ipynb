**Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.**

**Answer**

In Pandas, a DataFrame is a two-dimensional tabular data structure with rows and columns, where each column can have a different data type (e.g., integer, float, string, etc.). A DataFrame is the most commonly used data structure in Pandas.

On the other hand, a pandas.Series is a one-dimensional labeled array that can hold data of any type (e.g., integer, float, string, etc.). It can be thought of as a single column of a spreadsheet. A Series has an index that allows for fast and efficient data retrieval.

```python
# print the DataFrame
print(my_data)

# create a Series from the 'name' column of the DataFrame
name_series = my_data['name']

# print the Series
print(name_series)
```

**Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can you give an example of when you might use one of these functions?**

**Answer:**

There are many common functions you can use to manipulate data in a Pandas DataFrame. Here are some examples:

drop(): to drop rows or columns from a DataFrame based on a specified axis and labels.

fillna(): to fill missing values in a DataFrame with a specified value or method. 

groupby(): to group rows in a DataFrame based on one or more columns. 

merge(): to merge two or more DataFrames based on a common key. 

sort_values(): to sort a DataFrame by one or more columns. 

pivot_table(): to create a spreadsheet-style pivot table based on a DataFrame. 

**Example:**
You might use pivot_table() function to summarize data by category and then compare the results across categories and sort_values() function to sort a DataFrame by date or by a numeric column in ascending or descending order.

**Q7.Create a DataFrame using multiple Series. Explain with an example.**

```python
import pandas as pd

make = pd.Series(['Honda', 'Toyota', 'Ford', 'Chevrolet', 'BMW'])
model = pd.Series(['Civic', 'Camry', 'Mustang', 'Cruze', 'M3'])
year = pd.Series([2018, 2019, 2017, 2018, 2020])
price = pd.Series([15000, 20000, 18000, 17000, 40000])
```

```python
cars = pd.concat([make, model, year, price], axis=1, keys=['Make', 'Model', 'Year', 'Price'])
cars.head()
```

we have created a DataFrame called cars by concatenating the make, model, year, and price Series along the columns axis. We have specified the column names using the keys parameter, which creates a dictionary-like mapping between the keys and the Series.
