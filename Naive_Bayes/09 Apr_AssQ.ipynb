**Q1. What is Bayes' theorem?**

**Answer:**

Bayes' theorem is a fundamental concept in probability theory that describes the probability of an event based on prior knowledge or observations.

The theorem states that the probability of a hypothesis (H) given some observed evidence (E) is proportional to the probability of the evidence given the hypothesis times the prior probability of the hypothesis.



**Q2. What is the formula for Bayes' theorem?**

**Answer:**

Mathematically, this can be expressed as:

P(H|E) = P(E|H) * P(H) / P(E)

where 

P(H|E) is the probability of hypothesis H given evidence E, 

P(E|H) is the probability of observing evidence E given that hypothesis H is true, 

P(H) is the prior probability of hypothesis H, and 

P(E) is the probability of observing evidence E regardless of the hypothesis.

**Answer:**

Conditional probability is the probability of an event A occurring, given that another event B has occurred. It can be written as P(A|B), and it represents the updated probability of A based on the occurrence of B.

Bayes' theorem is a formula that allows us to calculate the conditional probability of an event A given another event B, taking into account prior beliefs about the probability of A and the likelihood of observing B given that A is true.

= (4/10) * (3/10) * (1/2) = 0.06

P(Class=B|X1=3,X2=4) = P(X1=3|Class=B) * P(X2=4|Class=B) * P(Class=B)

= (1/7) * (3/7) * (1/2) = 0.018
