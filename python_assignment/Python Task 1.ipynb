**Q1.** Create a function which will take a list as an argument and return the product of all the numbers
after creating a flat list.
Use the below-given list as an argument for your function.
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']
Note: you must extract numeric keys and values of the dictionary also.

```python
from functools import reduce
from collections.abc import Iterable  

given_list = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science'], 'Machine Learning']

def product_of_number(element):
    def number_list(element):
        num_list = []

        if isinstance(element, Iterable) and isinstance(element, dict):
            for k, v in element.items():
                num1 = number_list(k)
                num_list.extend(num1)
                num2 = number_list(v)
                num_list.extend(num2)

        elif isinstance(element, Iterable) and not isinstance(element, str):
            for i in element:
                num = number_list(i)
                num_list.extend(num)

        else:
            if type(element) == int:
                num_list.append(element)
        return num_list
    
    num_list = number_list(element)
    final_answer = reduce((lambda x, y: x * y), num_list)

    return num_list, final_answer

num_list, final_answer = product_of_number(given_list)
print('The list of number is:', num_list)
print('Final product is:', final_answer)
```

**Q2.** Write a python program for encrypting a message sent to you by your friend. The logic of encryption should be such that, for a the output should be z. For b, the output should be y. For c, the output should be x respectively. Also, the whitespace should be replaced with a dollar sign. Keep the punctuation
marks unchanged.

Input Sentence: I want to become a Data Scientist.

Encrypt the above input sentence using the program you just created.
Note: Convert the given input sentence into lowercase before encrypting. The final output should be
lowercase.

```python
message = 'I want to become a Data Scientist.'
import string

def encrypt_message(message):
    message = message.lower()
    alphabet_list = list(string.ascii_lowercase)
    encrypted_message = ''

    for word in message:
        if word in alphabet_list:
            idx = alphabet_list.index(word)
            encrypted_message += alphabet_list[len(alphabet_list) - idx - 1]
        elif word == ' ':
            encrypted_message += '$'
        else:
            encrypted_message += word

    return encrypted_message

encrypted_message = encrypt_message(message)
print(encrypt_message(message))
```
