**Q1. What is an exception in python? Write the difference between Exceptions and Syntax errors?**

During the execution of a program if an event or code disrupts the normal flow of the program's instructions, that event is known as exception. In general, when a Python script encounters a situation which cannot be handled, it raises an exception and it is an object that represents an error.

Syntax errors are the errors which is raised when there is an error in Python syntax whereas exception can also be a logical error.

**Q2. What happens when an exception is not handled? Explain with an example.**

If there is an exception while running the program, python will raise an error based on the error encountered. 

For instance, 

- try to open a file(for reading) that does not exist (FileNotFoundError)

- try to divide a number by zero (ZeroDivisionError)

- try to import a module that does not exist (ImportError) and so on.


**Q3. Which python statements are used to catch and handle exceptions? Explain with an example.**

Exceptions in python can be handled with the help of try and except statements. 

In try and except statement, try will first try to execute the desired code, if code run OK, the program will continue to run otherwise, to program will move to except block and run the code written in this block. 

```python
# example 
try:
    file = open('text.txt', 'r')
except Exception as e:
    print('The file does not exist')
```

**Q4. Explain with an example:**

**- try and else**

**- finally**

**- raise** 

```python
# Try and Else
'''
Try and else statement is a condition is python which is excecuted togather. Else will execute only when try is executed otherwise it won't execute.
'''
# Example 
try:
    print('This is a try block')
except Exception as e:
    print('There is a problem')
else:
    print('Try has been executed')
```

```python
# finally
'''
finally is a part of exception handling in try and except statements. block of code inside finally statement will always execute irrespective to error or not.
'''

# example
try:
    a = int(input())
except Exception as e:
    print('There is a problem')
finally:
    print('Try has been executed')
```

```python
# raise
'''
The raise keyword is used to raise an exception.
You can define what kind of error to raise, and the text to print to the user.
'''

# example
x = "hello"

if not type(x) is int:
  raise TypeError("Only integers are allowed")
```

**Q5. What are Custom Exceptions in python? Why do we need Custome Exceptions? Explain with an example.**

in python, sometimes we may need to create our own custom exceptions that serve our purpose. Custom exception are similar to python exception but it is customise to raise a specific information as per user requirements. 

We can define custom exceptions by creating a new class that is derived from the built-in Exception class.

For instance, if we want user to enter his or her age. We can raise custom excpetion based on pre-defined criterion such as age must be a positive integer and should not be greater than 150 etc. 

**Q6. Create a custom exception class. use this class to handle an exception.**

```python
# example 

# define user-defined exceptions
class InvalidAgeException(Exception):
    "Raised when the input value is less than 18"
    pass

# you need to guess this number
age = 18

try:
    input_num = int(input("Enter a number: "))
    if input_num < age:
        raise InvalidAgeException
    else:
        print("Eligible to Vote")
        
except InvalidAgeException:
    print("Exception occurred: Invalid Age")
```
