**Q1. What is an ensemble technique in machine learning?**

**Answer:**

In machine learning, ensemble techniques refer to the methods of combining multiple models to improve the accuracy of predictions. The idea is to build multiple models that are different from each other in some way, and then combine their predictions to get a more accurate result than any individual model. 

The main advantages of ensemble techniques are:

**Increased accuracy:** Ensemble techniques can significantly increase the accuracy of predictions compared to a single model. This is because the ensemble model is able to capture more complex patterns and relationships in the data.

**Robustness:** Ensemble techniques are more robust to overfitting and noise in the data. This is because the ensemble model is able to average out the errors and inconsistencies in the individual models.

**Q7. How is the confidence interval calculated using bootstrap?**

**Answer:**

Steps to calculate the confidence interval using bootstrap:

Choose the statistic of interest (e.g., mean, median, standard deviation) and calculate it for the original dataset.

Create multiple bootstrap samples by randomly sampling with replacement from the original dataset. Each bootstrap sample should have the same size as the original dataset.

Calculate the statistic of interest for each bootstrap sample.

Calculate the mean and standard deviation of the statistic across the bootstrap samples.

Use the mean and standard deviation to construct the confidence interval. A common approach is to use the percentile method, which involves finding the values of the statistic at the lower and upper percentiles of the distribution.


**Q8. How does bootstrap work and What are the steps involved in bootstrap?**

**Answer:**

Bootstrap is a resampling technique that involves creating new samples by randomly sampling with replacement from the original dataset. It is used to estimate the sampling distribution of a statistic or to calculate the confidence interval of a parameter.

Here are the steps involved in bootstrap:

**Choose the sample size:** Determine the sample size that you want to use for each bootstrap sample. The sample size should be equal to the size of the original dataset.

**Create bootstrap samples:** Create multiple bootstrap samples by randomly sampling with replacement from the original dataset. Each bootstrap sample should have the same size as the sample size you chose in step 1.

**Calculate the statistic of interest:** Calculate the statistic of interest for each bootstrap sample. This could be a mean, median, variance, standard deviation, or any other statistic that you are interested in.

**Estimate the sampling distribution:** Use the distribution of the statistic across the bootstrap samples to estimate the sampling distribution of the statistic. This can be done by calculating the mean and standard deviation of the statistic across the bootstrap samples.

**Calculate the confidence interval:** Use the sampling distribution to calculate the confidence interval of the statistic. A common approach is to use the percentile method, which involves finding the values of the statistic at the lower and upper percentiles of the distribution.

**Repeat steps 2-5:** Repeat steps 2-5 multiple times to obtain more accurate estimates of the sampling distribution and confidence interval.


**Q9. A researcher wants to estimate the mean height of a population of trees. They measure the height of a sample of 50 trees and obtain a mean height of 15 meters and a standard deviation of 2 meters. Use bootstrap to estimate the 95% confidence interval for the population mean height.**

```python
import numpy as np

# Define the sample data
sample_heights = np.random.normal(loc=15, scale=2, size=50)

# Define the number of bootstrap samples to take
n_bootstrap = 10000

# Initialize an array to hold the bootstrap sample means
bootstrap_means = np.empty(n_bootstrap)

# Draw bootstrap samples and calculate means
for i in range(n_bootstrap):
    bootstrap_sample = np.random.choice(sample_heights, size=50, replace=True)
    bootstrap_means[i] = np.mean(bootstrap_sample)

# Calculate the standard error of the mean
se = np.std(bootstrap_means)

# Calculate the confidence interval
ci_lower = np.mean(sample_heights) - 1.96 * se
ci_upper = np.mean(sample_heights) + 1.96 * se

print("95% confidence interval for population mean height:", ci_lower, "-", ci_upper)
```
