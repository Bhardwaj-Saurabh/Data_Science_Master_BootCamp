**Answer:**

For a regression problem such as predicting house prices based on characteristics, the most commonly used evaluation metric is Mean Squared Error (MSE). This metric measures the average squared difference between the predicted and actual values, which gives a sense of how far off the model's predictions are from the true values. Additionally, other metrics such as Root Mean Squared Error (RMSE) and R-squared can also be used to evaluate the performance of the SVM regression model.




**Q2. You have built an SVM regression model and are trying to decide between using MSE or R-squared as your evaluation metric. Which metric would be more appropriate if your goal is to predict the actual price of a house as accurately as possible?**

**Answer:**

If the goal is to predict the actual price of a house as accurately as possible, then Mean Squared Error (MSE) would be a more appropriate evaluation metric than R-squared.

This is because MSE measures the average squared difference between the predicted and actual values, giving more weight to larger differences. Therefore, if the goal is to minimize the error in predicting the actual price of a house, MSE would be a better metric to use as it penalizes larger errors more heavily.

**Q3. You have a dataset with a significant number of outliers and are trying to select an appropriate regression metric to use with your SVM model. Which metric would be the most appropriate in this scenario?**

**Answer:** 

When dealing with a dataset that contains a significant number of outliers, mean squared error (MSE) might not be the most appropriate metric to evaluate the performance of a Support Vector Regression (SVR) model. This is because MSE gives equal weight to all errors, regardless of their magnitude.

Instead, a more appropriate metric for evaluating the performance of an SVR model in the presence of outliers would be mean absolute error (MAE). Unlike MSE, MAE places equal weight on all errors, making it less sensitive to the impact of outliers.
